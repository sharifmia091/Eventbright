```python
import requests, os, random, datetime, re,datetime,time,os
from bs4 import BeautifulSoup
import pandas as pd
```

```python
event_list =[]
```

```python
def pg_scrape(url):
    website=requests.get(url)
    bs=BeautifulSoup(website.content,'html.parser')
    all_events=bs.find('ul', {'class':'search-main-content__events-list'})
    events= all_events.find_all('li')
    
    
    
    for event in events:
        event_dict={}
        name_tag=event.find('div',{'class':'eds-event-card__formatted-name--is-clamped'})
        name= name_tag.text
        event_dict['name'] =name
        


        link= event.find('a').attrs['href']
        event_dict['link'] =link

        try:
            date=(event.find('div',{'class':'eds-text-bs--fixed eds-text-color--grey-600 eds-l-mar-top-1'}).text)
            event_dict['date'] =date
        except:
            event_dict['date'] ='N/A'
            continue
            
            


        loc=(event.find('div',{'class':'card-text--truncated__one'}).text)
        event_dict['loc'] =loc


        ws=requests.get(link)
        bsoup=BeautifulSoup(ws.content,'html.parser')
        
        try:
            article=bsoup.find('div',{'class':'g-cell g-cell-1-1 g-cell-md-7-12 g-cell--no-gutters l-lg-pad-right-6'}).find('strong')
            title=article.text 
            event_dict['title'] =title
        
        except:
            event_dict['title'] ='N/A'
            continue


        try:

            paragraph_tag= bsoup.find('div',{'class':'structured-content-rich-text'}).find('p')
            paragraph =paragraph_tag.text 
            event_dict['paragraph'] =paragraph

        except:
            
            event_dict['paragraph'] ='N/A'
            continue


        try:

            time= [i.text for i in bsoup.find('time',{'class':'clrfix'}).find_all('p')[0:2]] 
            event_dict['time'] =time
        
        except:
            
            event_dict['time']  ='N/A'
            continue
        
        event_list.append(event_dict)
 
    
    return 
        
        
```

```python
count = 1
while count<50:
    url="https://www.eventbrite.com.au/d/australia/all-events/?page='+str(count)+'/'"
    pg_scrape(url)
    print('https://www.eventbrite.com.au/d/australia/all-events/?page='+str(count)+'/')
    time.sleep(1)
    count+=1
    
```

```python
df=pd.DataFrame(event_list)
```
