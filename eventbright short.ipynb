```python
import requests, os, random, datetime, re,datetime,time,os
from bs4 import BeautifulSoup
import pandas as pd
```

```python
event_list =[]
```

```python
def pg_scrape(url):
    website=requests.get(url)
    bs=BeautifulSoup(website.content,'html.parser')
    all_events=bs.find('ul', {'class':'search-main-content__events-list'})
    events= all_events.find_all('li')
    
    for event in events:
    
        event_dict={}
        name_tag=event.find('div',{'class':'eds-event-card__formatted-name--is-clamped'})
        name= name_tag.text
        event_dict['name'] =name



        link= event.find('a').attrs['href']
        event_dict['link'] =link


        date=(event.find('div',{'class':'eds-text-bs--fixed eds-text-color--grey-600 eds-l-mar-top-1'}).text)
        event_dict['date'] =date

        loc=(event.find('div',{'class':'card-text--truncated__one'}).text)
        event_dict['loc'] =loc
        
        event_list.append(event_dict)
    
    return 
        
        
```

```python
count = 1
while count<50:
    url="https://www.eventbrite.com.au/d/australia/all-events/?page='+str(count)+'/'"
    pg_scrape(url)
    
    print('https://www.eventbrite.com.au/d/australia/all-events/?page='+str(count)+'/')
    time.sleep(5)
    count+=1
    
```

```python
df=pd.DataFrame(event_list)
```
